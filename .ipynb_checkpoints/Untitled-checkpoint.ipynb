```python
# Crear una fila de etiquetas para las muestras
labels_row = pd.Series(data.columns.map(class_mapping), index=data.columns, name='Label')
# Convertir la serie en un DataFrame de una sola fila
labels_df = labels_row.to_frame().T
# Cambiar el nombre del índice de la fila para evitar conflictos
labels_df.index = ['Labels']
# Añadir la fila de etiquetas al DataFrame `data`
data = pd.concat([data, labels_df], sort=False)
data = data.T
data
```

```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Separar la columna 'Labels' de los datos numéricos
labels = data['Labels']  # Asumiendo que la columna se llama 'Labels'
data_numeric = data.drop(columns=['Labels'])

# Convertir todas las columnas a tipo numérico, forzando a NaN los valores que no se puedan convertir
data_numeric = data_numeric.apply(pd.to_numeric, errors='coerce')

# Asegurarse de que no haya valores NaN
data_numeric = data_numeric.dropna(axis=1, how='any')

# Paso 1: Calcular la varianza de cada gen y seleccionar los 500 genes más variables
variance = data_numeric.var(axis=0)
top_500_genes = variance.nlargest(500).index
data_top_500 = data_numeric[top_500_genes]

# Paso 2: Estandarizar los datos de los 500 genes más variables
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_top_500)

# Paso 3: Calcular la matriz de distancias y realizar el clustering jerárquico
Z = linkage(data_scaled, method='ward')  # Puedes cambiar 'ward' por 'single', 'complete', etc.

# Paso 4: Visualizar el dendrograma
plt.figure(figsize=(10, 7))
dendrogram(Z, labels=labels.values, leaf_rotation=90, leaf_font_size=10)
plt.title('Dendrograma del Clustering Jerárquico')
plt.xlabel('Muestras')
plt.ylabel('Distancia')
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Paso 1: Realizar PCA en los datos escalados
pca = PCA()
pca.fit(data_scaled)

# Paso 2: Evaluar la varianza explicada acumulada
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = explained_variance_ratio.cumsum()

# Paso 3: Graficar la varianza explicada acumulada
plt.figure(figsize=(10, 6))
plt.plot(cumulative_variance, marker='o')
plt.axhline(y=0.70, color='r', linestyle='--')  # Línea de referencia para el 90% de varianza explicada
plt.title('Varianza Explicada Acumulada por Componentes Principales')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Explicada Acumulada')
plt.show()

# Número de componentes necesarios para alcanzar el umbral del 90% de varianza explicada
threshold = 0.70
num_components = (cumulative_variance >= threshold).argmax() + 1
print(f'Número de componentes necesarios para alcanzar el {threshold*100}% de varianza explicada: {num_components}')
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Paso 1: Seleccionar los 50 genes más variables
top_50_genes = data_numeric.var(axis=0).nlargest(50).index
data_top_50 = data_numeric[top_50_genes]

# Paso 2: Estandarizar los datos de los 50 genes más variables
scaler = StandardScaler()
data_top_50_scaled = scaler.fit_transform(data_top_50)

# Paso 3: Realizar PCA en los 50 genes más variables
pca = PCA()
pca.fit(data_top_50_scaled)

# Paso 4: Evaluar la varianza explicada acumulada
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = explained_variance_ratio.cumsum()

# Paso 5: Graficar la varianza explicada acumulada
plt.figure(figsize=(10, 6))
plt.plot(cumulative_variance, marker='o')
plt.axhline(y=0.70, color='r', linestyle='--')  # Línea de referencia para el 90% de varianza explicada
plt.title('Varianza Explicada Acumulada por Componentes Principales (50 Genes Más Variables)')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Explicada Acumulada')
plt.show()

# Número de componentes necesarios para alcanzar el umbral del 90% de varianza explicada
threshold = 0.70
num_components = (cumulative_variance >= threshold).argmax() + 1
print(f'Número de componentes necesarios para alcanzar el {threshold*100}% de varianza explicada: {num_components}')
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Paso 1: Aplicar PCA con 5 componentes principales
pca = PCA(n_components=3)
data_pca_5 = pca.fit_transform(data_top_50_scaled)

# Paso 2: Determinar el número óptimo de clusters usando el método del codo
inertia = []
for n_clusters in range(2, 11):  # Probar entre 2 y 10 clusters
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(data_pca_5)
    inertia.append(kmeans.inertia_)

# Graficar el método del codo
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), inertia, marker='o')
plt.title('Método del Codo para Determinar el Número Óptimo de Clusters')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.show()

```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from kneed import KneeLocator

# Paso 1: Aplicar PCA con 3 componentes principales (ajustar si es necesario)
pca = PCA(n_components=3)
data_pca_5 = pca.fit_transform(data_top_50_scaled)

# Paso 2: Determinar el número óptimo de clusters usando el método del codo
inertia = []
for n_clusters in range(2, 11):  # Probar entre 2 y 10 clusters
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(data_pca_5)
    inertia.append(kmeans.inertia_)

# Encontrar el codo usando kneed
kneedle = KneeLocator(range(2, 11), inertia, curve='convex', direction='decreasing')
optimal_clusters = kneedle.elbow

# Graficar el método del codo
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), inertia, marker='o', label='Inercia')
plt.axvline(x=optimal_clusters, color='r', linestyle='--', label=f'Óptimo Clusters: {optimal_clusters}')
plt.title('Método del Codo para Determinar el Número Óptimo de Clusters')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.legend()
plt.show()

print(f'Número óptimo de clusters según el método del codo: {optimal_clusters}')
```

```python
# Paso 3: Aplicar K-means con el número óptimo de clusters (ajustar según el método del codo)
optimal_clusters = 4 # Ajusta según el método del codo
kmeans = KMeans(n_clusters=optimal_clusters, random_state=40)
clusters = kmeans.fit_predict(data_pca_5)

# Añadir los resultados al DataFrame original
data['KMeans_Cluster'] = clusters
```

```python
# Visualizar los resultados de clustering en los primeros dos componentes principales con etiquetas
plt.figure(figsize=(10, 7))
scatter = plt.scatter(data_pca_5[:, 0], data_pca_5[:, 1], c=clusters, cmap='viridis', marker='o')


# Añadir etiquetas a los puntos
for i, label in enumerate(data['Labels']):  # Usa la columna 'Labels' como etiquetas
    plt.annotate(label, (data_pca_5[i, 0], data_pca_5[i, 1]), fontsize=8, alpha=0.7)

plt.title('K-means Clustering en los Primeros dos Componentes Principales')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.colorbar(scatter, label='Cluster')
plt.show()
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# DataFrame con los datos de los 50 genes más variables, etiquetas y clusters
data_top_50_with_labels_and_clusters = data_top_50.copy()
data_top_50_with_labels_and_clusters['Labels'] = data['Labels']
data_top_50_with_labels_and_clusters['KMeans_Cluster'] = data['KMeans_Cluster']

# Crear un DataFrame para almacenar los resultados
results_list = []

# Iterar sobre todos los clusters
for cluster in data_top_50_with_labels_and_clusters['KMeans_Cluster'].unique():
    # Filtrar el DataFrame para el cluster de interés
    data_cluster = data_top_50_with_labels_and_clusters[data_top_50_with_labels_and_clusters['KMeans_Cluster'] == cluster]

    # Separar las muestras por etiquetas dentro del cluster
    ac_samples = data_cluster[data_cluster['Labels'] == 'AC']
    scc_samples = data_cluster[data_cluster['Labels'] == 'SCC']

    # Verificar que haya muestras suficientes para cada etiqueta
    if len(ac_samples) > 1 and len(scc_samples) > 1:
        # Eliminar las columnas de etiquetas y clusters antes de realizar el análisis
        ac_samples = ac_samples.drop(columns=['Labels', 'KMeans_Cluster'])
        scc_samples = scc_samples.drop(columns=['Labels', 'KMeans_Cluster'])

        # Calcular la media de expresión para cada gen en cada grupo
        mean_ac = ac_samples.mean()
        mean_scc = scc_samples.mean()

        # Realizar la prueba t para cada gen entre los dos grupos
        p_values = {}
        for gene in ac_samples.columns:
            _, p_value = ttest_ind(ac_samples[gene], scc_samples[gene], nan_policy='omit')
            p_values[gene] = p_value

        # Crear un DataFrame con los resultados
        results = pd.DataFrame({
            'Cluster': cluster,
            'Mean_AC': mean_ac,
            'Mean_SCC': mean_scc,
            'P_Value': pd.Series(p_values)
        })

        # Filtrar genes con valores p significativos (ajustar umbral según sea necesario)
        significant_genes = results[results['P_Value'] < 0.05]  # Ajustar el umbral de significancia

        # Ordenar los genes por la diferencia en expresión media
        significant_genes['Expression_Difference'] = abs(significant_genes['Mean_AC'] - significant_genes['Mean_SCC'])
        significant_genes = significant_genes.sort_values(by='Expression_Difference', ascending=False)

        # Añadir resultados a la lista
        results_list.append(significant_genes)

# Concatenar los resultados de todos los clusters en un solo DataFrame
final_results = pd.concat(results_list)

# Mostrar los resultados
final_results
```
